#2.	What factors are significantly related to the number of web purchases?


Strong Positive Correlations:

MntWines (0.542): People who purchase more wine online also tend to make more web purchases. 
This suggests a potential link between high wine spending and frequent online shopping.
    
NumStorePurchases (0.503): There is a moderate to strong positive correlation, indicating that individuals who make more purchases in-store also tend to make more purchases online.
MntGoldProds (0.422): Spending on gold products is also positively correlated with online purchases. Those who spend more on gold tend to buy more online.
Income (0.388): Higher income is associated with a higher number of web purchases. This makes sense as individuals with higher incomes may have more disposable income to spend online.
NumCatalogPurchases (0.378): More catalog purchases are linked with more web purchases, suggesting that those who engage with catalogs may also shop online frequently.

Moderate Positive Correlations:

MntSweetProducts (0.349): Spending on sweet products is positively related to web purchases. Customers who buy sweet products might also be more inclined to shop online.
MntFruits (0.297), MntMeatProducts (0.294), MntFishProducts (0.294): Spending on these food categories shows moderate positive correlations with online purchases, indicating that those who spend on these items may also engage in online shopping.

Low Positive Correlations:

AcceptedCmp4 (0.156), Teenhome (0.156), AcceptedCmp1 (0.155): These have lower positive correlations, suggesting they have a slight impact on web purchases.
Response (0.149), Age (0.145): The age and response to campaigns show small positive correlations with web purchases, indicating they have some influence but not a strong one.

Negative Correlations:

Kidhome (-0.362): Having children at home shows a significant negative correlation with web purchases. This could imply that individuals with children might spend less on online shopping, possibly due to other priorities or less discretionary income.

Education_Basic (-0.124), Year_Birth (-0.145): Basic education and older birth year have a slight negative correlation, suggesting that those with basic education or older age might be less likely to make online purchases.
Country-Specific Variables (e.g., Country_Spain (-0.029), Country_India (-0.013)): The correlations with specific countries are generally weak, suggesting that the country of residence has a minor impact on web purchase behavior.



# Key Takeaways:
High-spending categories like wines and gold products are closely tied to online purchasing behavior, indicating that these customer segments are strong targets for e-commerce campaigns.

Income is a significant factor in online purchasing, making wealthier individuals important prospects for online retail marketing.

Multi-channel shoppers, who shop both in-store and through catalogs, are also frequent online buyers, indicating opportunities for cross-channel promotions.

Parents and those with basic education or older age are less likely to shop online, which can inform more targeted marketing strategies aimed at boosting online engagement within these segments.

These insights highlight the importance of focusing on high-income, high-spending segments and leveraging multi-channel shopping habits while recognizing the limitations in certain demographics like families with children and older individuals.

```python

# Assuming df is your dataframe
# Classify columns into 'Monetary' and 'Frequency'
monetary_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
frequency_columns = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']

# Unpivot (melt) monetary columns
df_monetary = pd.melt(df, id_vars=['ID'], value_vars=monetary_columns, 
                      var_name='MonetaryType', value_name='AmountSpent')

# Unpivot (melt) frequency columns
df_frequency = pd.melt(df, id_vars=['ID'], value_vars=frequency_columns, 
                       var_name='PurchaseChannel', value_name='PurchaseCount')

# Display the unpivoted dataframes
print(df_monetary.head())   # Shows unpivoted monetary columns
print(df_frequency.head())  # Shows unpivoted frequency columns
```

```python
df_frequency
```

```python
sns.set(style="whitegrid")

#Barplot for Frequency data
plt.figure(figsize=(10, 6))
sns.barplot(data=df_frequency, x='PurchaseChannel', y='PurchaseCount', ci=None)
plt.title('Purchase Channel by Purchase Count')
plt.xticks(rotation=45)
plt.show()

# Store purchased has the highest channel in all.
```

```python

#group by with ID
df_monetary = df_monetary.groupby('ID')['AmountSpent'].sum().reset_index()  

df_monetary.describe()

df_frequency = df_frequency.groupby('ID')['PurchaseCount'].sum().reset_index()
df_frequency.describe()
```

```python
# Plot 3: Boxplot to show the distribution of spending for different product types
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_monetary, x='MonetaryType', y='AmountSpent')
plt.title('Spending Distribution by Product Type')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 15))
plt.subplot(3, 3, 1)
sns.distplot(df_selected['Recency'])
plt.subplot(3, 3, 2)
sns.distplot(df_selected['PurchaseCount'])
plt.subplot(3, 3, 3)
sns.distplot(df_selected['AmountSpent'])
```

```python



# Define bins
r_bins = [-1, 24, 74, 99]
f_bins = [0, 8, 21, 44]
m_bins = [5, 68, 1045, 2525]

# Assign scores
df_selected.loc[:,'r_score'] = pd.cut(df_selected['Recency'], r_bins, labels=[3, 2, 1])
df_selected.loc[:,'f_score'] = pd.cut(df_selected['PurchaseCount'], f_bins, labels=[1, 2, 3])
df_selected.loc[:,'m_score'] = pd.cut(df_selected['AmountSpent'], m_bins, labels=[1, 2, 3])

# Convert scores to integers
df_selected.loc[:,'r_score'] = df_selected['r_score'].fillna(0).astype(int)
df_selected.loc[:,'f_score'] = df_selected['f_score'].fillna(0).astype(int)
df_selected.loc[:,'m_score'] = df_selected['m_score'].fillna(0).astype(int)

# Create the rfm_group column
df_selected.loc[:,'rfm_group'] = df_selected['r_score'].astype(str) + df_selected['f_score'].astype(str) + df_selected['m_score'].astype(str)

# Display the DataFrame
print(df_selected)

```

```python
# Set the style for seaborn plots
sns.set(style="whitegrid")

# Create a figure for a single plot
plt.figure(figsize=(10, 6))

# Histogram for RFM Group
sns.histplot(df_selected['rfm_group'].astype(str), bins=len(df_selected['rfm_group'].unique()), kde=False)
plt.title('Histogram of RFM Group Distribution')
plt.xlabel('RFM Group')
plt.ylabel('Frequency')
plt.xticks(rotation=45)


# Adjust layout
plt.tight_layout()
plt.show()
```
